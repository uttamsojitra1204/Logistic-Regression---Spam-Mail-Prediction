```python
#import the neccesary libarary

import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")
```

```python
#collection the data and preprocessing

df= pd.read_csv("mail_data.csv")
df.head()
```

```python
#how many rows and columns are there

df.shape
```

```python
#check null value or not

df.isnull().sum()
```

```python
#label encoding
#0 => spam
#1 => ham

df.loc[df["Category"] == "spam", "Category",] = 0
df.loc[df["Category"] == "ham", "Category",] = 1
```

```python
df.head(3)
```

```python
#separating the data x and y

X= df["Message"]
Y= df["Category"]
```

```python
print(X)
```

```python
print(Y)
```

```python
#splitting the data train and test

X_train, X_test, Y_train, Y_test= train_test_split(X, Y, random_state= 42, test_size= 0.2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#transfrom the text data to features vector that can be used as input to the logistic regression

feature_extraction= TfidfVectorizer(min_df= 1, stop_words= "english", lowercase= True)

X_train_feature= feature_extraction.fit_transform(X_train)
X_test_feature= feature_extraction.transform(X_test)

#convert Y_train and Y_test valur into the integer
Y_train= Y_train.astype("int")
Y_test= Y_test.astype("int")
```

```python
print(X_train)
```

```python
print(X_train_feature)
```

```python
#train the model

lr= LogisticRegression()
lr.fit(X_train_feature, Y_train)
```

```python
#evaluate the model
```

```python
#prediction on train data
train_pred= lr.predict(X_train_feature)
train_acc= accuracy_score(train_pred, Y_train)
print(train_acc)
```

```python
#prediction on test data
test_pred= lr.predict(X_test_feature)
test_acc= accuracy_score(test_pred, Y_test)
print(test_acc)
```

```python
#building prediction system
input= ["it is free. join and get rewads."]

#convert text to feature vector
input_feature= feature_extraction.transform(input)

pred= lr.predict(input_feature)
print(pred)

if pred[0] == 1:
    print("Mail is ham")
else:
    print("Mail is Spam")
```
